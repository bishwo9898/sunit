```python
file = open('DSC270_WSL_Data.csv', 'r')
line = file.readline()
yellow_list = []
red_list = []

while line != "":
    line = file.readline()
    line = line.strip()
    row = line.split("\t")
    if len(row) > 12:
        name = row[1].strip()
        yellow = int(row[11])
        red = int(row[12])
    
    if yellow >= 5:
        yellow_list.append(name)

    if red >= 1:
        red_list.append(name)
        
print("Name of players who got yellow card: ")
for names in yellow_list:
    print(names)
print("\nYellow cards count: ", len(yellow_list))

print("\nName of players who got red card: ")

print("\n")
for names in red_list:
    print(names)

```

```python
file = open('DSC270_WSL_Data.csv', 'r')
line = file.readline()
goals_list = []
pk_list = []
pk_attempts_list = []

while line != "":
    line = file.readline()
    line = line.strip()
    row = line.split("\t")
    if len(row) > 12:
        #name = row[1].strip()
        goal = int(row[8])
        pk = int(row[9])
        pk_attempts = int(row[10])
        pk_list.append(pk)
        pk_attempts_list.append(pk_attempts)
        goals_list.append(goal)


k = max(goals_list)
#g = k - pk
print("Most goals scored by a player: ", k)
```

```python
missed_list = []  

file = open('DSC270_WSL_Data.csv', 'r')
line = file.readline()

while line != "":
    line = file.readline()
    line = line.strip()
    row = line.split("\t")
    if len(row) > 12:
        pk = int(row[9])
        pk_attempts = int(row[10])
        missed = pk_attempts - pk
        missed_list.append(missed)

most_missed = max(missed_list)
print("Most missed penalty kicks: ", most_missed)
```

```python
file = open('DSC270_WSL_Data.csv', 'r')
line = file.readline()
worst_pk_percent = 100  
worst_pk_players = []

while line != "":
    line = file.readline()
    line = line.strip()
    row = line.split("\t")
    if len(row) > 12:
        name = row[1].strip()
        pk_made = int(row[9])
        pk_attempts = int(row[10])
        
        if pk_attempts > 0:  
            pk_percent = (pk_made / pk_attempts) * 100
            
            if pk_percent < worst_pk_percent:
                worst_pk_percent = pk_percent
                worst_pk_players = []
                worst_pk_players.append(name)
            elif pk_percent == worst_pk_percent:
                worst_pk_players.append(name)

print("Player(s) with worst PK %: ")
for player in worst_pk_players:
    print(player)

print("\nWorst PK %: ", round(worst_pk_percent, 2))
```

```python
file = open('DSC270_WSL_Data.csv', 'r')
line = file.readline()
teams = []
team_goals = []

while line != "":
    line = file.readline()
    line = line.strip()
    row = line.split("\t")
    if len(row) > 12:
        team = row[4].strip()
        goals = int(row[5])
        if team in teams:
            index = teams.index(team)
            team_goals[index] += goals
        else:
            teams.append(team)
            team_goals.append(goals)

max_goals = 0
for g in team_goals:
    if g > max_goals:
        max_goals = g

for i in range(len(teams)):
    if team_goals[i] == max_goals:
        print(teams[i])

print("Total goals: ", max_goals)
```

```python
file = open('DSC270_WSL_Data.csv', 'r')
line = file.readline()

chelsea_players = []
chelsea_assists = []
gk_players = []

while line != "":
    line = file.readline()
    line = line.strip()
    row = line.split("\t")
    if len(row) > 12:
        name = row[1].strip()
        team = row[4].strip()
        assists = int(row[6])
        position = row[3].strip()
        
        if team == "Chelsea":
            chelsea_players.append(name)
            chelsea_assists.append(assists)
        
        if position == "GK":
            gk_players.append(name)

max_assists = max(chelsea_assists)
print("Chelsea player(s) with most assists: ")
for i in range(len(chelsea_players)):
    if chelsea_assists[i] == max_assists:
        print(chelsea_players[i])

print("\nGoalkeepers in the league: ")
for player in gk_players:
    print(player)
```

```python
file = open('DSC270_WSL_Data.csv', 'r')
line = file.readline()

positions = []

while line != "":
    line = file.readline()
    line = line.strip()
    row = line.split("\t")
    if len(row) > 12:
        pos = row[3].strip()
        if pos not in positions:
            positions.append(pos)

positions.sort()
print("All positions in the league (sorted:)")
for pos in positions:
    print(pos)
```

```python
file = open('DSC270_WSL_Data.csv', 'r')
line = file.readline()

pos_count = {}

while line != "":
    line = file.readline()
    line = line.strip()
    row = line.split("\t")
    if len(row) > 12:
        pos = row[3].strip()
        if pos in pos_count:
            pos_count[pos] += 1
        else:
            pos_count[pos] = 1

print("Number of players per position:")
for position in pos_count:
    print(position, ":", pos_count[position])
```
